```python
from transformers import pipeline
from datasets import load_dataset
```

```python
# Load the dataset from Hugging Face Datasets library
raw_datasets = load_dataset("squad")
```

```python
qa = pipeline(
    "question-answering",
    model="distilbert-base-uncased-distilled-squad",
    tokenizer="distilbert-base-uncased-distilled-squad",
)
```

```python
# Take the first sample from the dataset
sample = raw_datasets["train"][0]

question = sample["question"]
context = sample["context"]

print("Question:", question)
print("Context:", context[:300], "...")  # show only first 300 chars for readability
```

```python
res = qa({
    "question": question,
    "context": context
})

print("Answer:", res["answer"])
```

```python
custom_context = """
The Stanford Question Answering Dataset (SQuAD) is a reading comprehension dataset consisting of questions posed 
by crowdworkers on a set of Wikipedia articles, where the answer to every question is a segment of text, or span, 
from the corresponding reading passage.
"""

custom_question = "What does SQuAD stand for?"

res_custom = qa({
    "question": custom_question,
    "context": custom_context
})

print("Custom Question:", custom_question)
print("Answer:", res_custom["answer"])
```
